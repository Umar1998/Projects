```python
from bs4 import BeautifulSoup 
import requests 
import csv
import pandas as pd
import time
from requests.adapters import HTTPAdapter
from requests.packages.urllib3.util.retry import Retry
import pandas as pd
```

```python
data_list = []
for i in range(45):
    req = requests.get("https://www.flipkart.com/mobiles/pr?sid=tyy%2C4io&marketplace=FLIPKART&page={}".format(i)).text
    soup = BeautifulSoup(req,'html.parser')
    material = soup.find_all('div', class_ = "_3pLy-c row")

    for x in material:

        data_dict= {}
        
        all_materials = x.find_all('li', class_ = "rgWa7D")
    
        try:
            name = x.find('div', class_ = "_4rR01T")
            data_dict["name"] = name.text
        except:
            data_dict["name"] = "not_listed"
                    
        try:
            rating = x.find('div', class_ = "_3LWZlK") 
            data_dict["rating"] = rating.text
        except:
            data_dict["rating"] = "not_listed"
            
        try:
            ram_and_rom = all_materials[0]
            data_dict["ram_and_rom"] = ram_and_rom.text
        except:
            data_dict["ram_and_rom"] = "not_listed"
        
        try:
            display = all_materials[1]
            data_dict["display"] = display.text
        except:
            data_dict["display"] = "not_listed"
                
        try:
            camera = all_materials[2]
            data_dict["camera"] = camera.text
        except:
            data_dict["camera"] = "not_listed"
                    
        try:
            battery = all_materials[3]
            data_dict["battery"] = battery.text
        except:
            data_dict["battery"] = "not_listed"
            
        try:
            processor = all_materials[4]
            data_dict["processor"] = processor.text
        except:
            data_dict["processor"] = "not_listed"

        try:
            warranty = all_materials[5]
            data_dict["warranty"] = warranty.text
        except:
            data_dict["warranty"] = "not_listed"

        try:
            price = x.find('div', class_ = "_30jeq3 _1_WHN1")
            data_dict["price"] = price.text
        except:
            data_dict["price"] = "not_listed"
            
        data_list.append(data_dict)
print(data_list)
```

```python
df = pd.DataFrame(data_list)
```
